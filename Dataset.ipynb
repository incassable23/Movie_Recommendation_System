```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
#using the uploaded csv files for datasets
```

```python
movies.head()
```

```python
movies.head(50)
```

```python
credits.head()
```

```python
total_movies = movies.merge(credits,on='title') #merging on the basis of title
```

```python
total_movies.shape #final merged dataset
```

```python
#genres
#id
#keywords
#title
#overview
#cast
#crew
#sorting the fields on the basis of which recommendation system will be formed

total_movies = total_movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
total_movies.head() #final data which was used in recommendation system
```

```python
#PREPROCESSING THE DATA
```

```python
#finding any null values 
total_movies.isnull().sum()
```

```python
#removing null values
total_movies.dropna(inplace=True)
total_movies.isnull().sum()
```

```python
#checking duplicate values
total_movies.duplicated().sum()
```

```python
#formatting the columns
import ast

def change(obj):
    List = []
    for i in ast.literal_eval(obj):
        List.append(i['name'])
    return List    
```

```python
total_movies['genres'] = total_movies['genres'].apply(change)
```

```python
total_movies['keywords'] = total_movies['keywords'].apply(change)
```

```python
def change1(obj):
    List = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            List.append(i['name'])
            counter+=1
        else:
            break
    return List 
```

```python
total_movies['cast'] = total_movies['cast'].apply(change1)
```

```python
total_movies.head()
```

```python
# for just having director as the crew value

def change2(obj):
    List = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            List.append(i['name'])
            break
    return List 
```

```python
total_movies['crew'] = total_movies['crew'].apply(change2)
```

```python
total_movies.head()
```

```python
#creating overview as list 
total_movies['overview'] = total_movies['overview'].apply(lambda x:x.split())

total_movies.head()
```

```python
#removing spaces so that no confusion arises in further searching outputs

total_movies['genres'] = total_movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
total_movies['keywords'] = total_movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
total_movies['cast'] = total_movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
total_movies['crew'] = total_movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
total_movies.head()
```

```python
total_movies['tags'] = total_movies['overview'] + total_movies['genres'] + total_movies['keywords'] + total_movies['cast'] + total_movies['crew']
```

```python
total_movies.head()
```

```python
final = total_movies[['movie_id','title','tags']]
final.head()
```

```python
#converting list into string

final['tags'] = final['tags'].apply(lambda x: " ".join(x))
final.head()
```

```python
#removing the similar words like actions which is same as action

import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#removing the similar words like actions which is same as action

def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
final['tags'] = final['tags'].apply(stem)
```

```python
final.head()
```

```python
#now apply text vectoristion so that related movies can be shown

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(final['tags']).toarray()
```

```python
vector
```

```python
cv.get_feature_names()
```

```python
#using the angles between vectors to calculate the similarities
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
def recommend(movie):
    movie_index = final[final['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(final.iloc[i[0]].title)
```

```python
recommend('Batman')
```
